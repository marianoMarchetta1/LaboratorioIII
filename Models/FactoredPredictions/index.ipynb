####  Diferencias 2018 - 2019 Por producto

Solo para los productos que tengan ventas en los meses de dic de 2018 y feb de 2019


```python
import pandas as pd

# Cargar el dataset
final_dataset = pd.read_csv('../../Datasets/final_dataset.csv', sep='\t')
final_dataset['periodo'] = pd.to_datetime(final_dataset['periodo'], format='%Y%m')

# Filtrar los datos para diciembre de 2018 y febrero de 2019
data_dec_2018 = final_dataset[(final_dataset['periodo'].dt.year == 2018) & (final_dataset['periodo'].dt.month == 12)]
data_feb_2019 = final_dataset[(final_dataset['periodo'].dt.year == 2019) & (final_dataset['periodo'].dt.month == 2)]

# Obtener los product_id con ventas en ambos periodos
product_ids_dec_2018 = data_dec_2018['product_id'].unique()
product_ids_feb_2019 = data_feb_2019['product_id'].unique()
common_product_ids = list(set(product_ids_dec_2018) & set(product_ids_feb_2019))

# Filtrar los datos solo para los product_id comunes
data_dec_2018_common = data_dec_2018[data_dec_2018['product_id'].isin(common_product_ids)]
data_feb_2019_common = data_feb_2019[data_feb_2019['product_id'].isin(common_product_ids)]

# Agrupar por product_id y calcular las ventas totales para diciembre de 2018 y febrero de 2019
ventas_dec_2018 = data_dec_2018_common.groupby('product_id')['y'].sum()
ventas_feb_2019 = data_feb_2019_common.groupby('product_id')['y'].sum()

# Calcular la variación porcentual de ventas entre diciembre de 2018 y febrero de 2019
variacion_porcentual = ((ventas_feb_2019 - ventas_dec_2018) / ventas_dec_2018) * 100

# Mostrar los resultados
for product_id, variacion in variacion_porcentual.items():
    print(f'Product {product_id}: Variación porcentual = {variacion:.2f}%')
```

#### Armo el dataset

Uso la dif de % para los productos que se pudo calcular, para los otros calculo el promedio

```python
data_dec_2019 = final_dataset[(final_dataset['periodo'].dt.year == 2019) & (final_dataset['periodo'].dt.month == 12)]
product_ids_dec_2019 = data_dec_2019['product_id'].unique()

pronosticos = []

for product_id in product_ids_dec_2019:
    if product_id in variacion_porcentual:
        # Calcular el pronóstico usando la variación porcentual
        variacion = variacion_porcentual[product_id]
        ventas_dec_2019 = data_dec_2019[data_dec_2019['product_id'] == product_id]['y'].sum()
        pronostico = ventas_dec_2019 * (1 + variacion / 100)
    else:
        # Calcular el pronóstico como el promedio de ventas para los meses disponibles
        ventas_promedio = final_dataset[final_dataset['product_id'] == product_id]['y'].mean()
        pronostico = ventas_promedio
    
    pronosticos.append({'product_id': product_id, 'predicted_y': pronostico})

# Crear un DataFrame con los pronósticos
pronosticos_df = pd.DataFrame(pronosticos)

# Guardar el DataFrame en un archivo CSV
pronosticos_df.to_csv('pronosticos_febrero_2020.csv', index=False)

# Mostrar el DataFrame con los pronósticos
print(pronosticos_df)
```
